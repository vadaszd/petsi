```python
from contextlib import contextmanager
import time

@contextmanager
def timing(activity: str = ""):
    start = time.time()
    yield
    elapsed = time.time() - start
    print(f"Duration of [{activity}]: {elapsed:0.1f} seconds")
```

```python
def plot_firing_interval(data, description, scale=(None, None), plot_gamma=True, plot_pareto=False, plot_kde=True, log_x=False, bins='auto', ):
    # https://stackoverflow.com/questions/15415455/plotting-probability-density-function-by-sample-with-matplotlib

    with timing("fitting a distribution"):
        dist_space = np.logspace(np.log10(min(data)) if min(data) > 0 else -100, np.log10(max(data)), 10) \
                        if log_x else linspace( min(data), max(data), 100 )

        if plot_kde:
            kde = gaussian_kde( data )    
            
            density = kde(dist_space)

        if plot_gamma:
            gamma_params = gamma.fit(data)
            print(f"gamma_params={gamma_params}")
            gamma_pdf = gamma.pdf(dist_space, *gamma_params)
            gamma_pdf[gamma_pdf > 1.0] = np.nan
    
        if plot_pareto:
            pareto_params = pareto.fit(data)
            print(f"pareto_params={pareto_params}")
            pareto_pdf = pareto.pdf(dist_space, *pareto_params)
            pareto_pdf[pareto_pdf > 1.0] = np.nan
    
    fig, ax = plt.subplots(figsize=(20, 6))
    # plt.figure
    plt.grid(True)
    
    with timing("creating a historgram"):
        ax.hist(data, bins=bins, density=True, label='Empirical density')    
        
    if plot_kde:
        ax.plot( dist_space, density, label='Estimated probability density (KDE)' )
    
    if plot_gamma:
        ax.plot( dist_space, gamma_pdf, label='Gamma distribution fit to the data'  )
        
    if plot_pareto:
        ax.plot( dist_space, pareto_pdf, label='Pareto distribution fit to the data'  )
        
    ax.legend()
    ax.text(0.98, 0.7, description, horizontalalignment='right', verticalalignment='center', multialignment='left',
            bbox=dict(boxstyle="round,pad=1", facecolor="w", edgecolor="0.5", ),
            transform=ax.transAxes)
    
    xscale, yscale = scale
    
    if xscale is not None:
        plt.xscale(xscale)
        
    if yscale is not None:
        plt.yscale(yscale)
        
    plt.show()
#     plt.figure(figsize=(20, 6))
#     plt.grid(True)
#     plt.loglog( dist_space, density )
```

```python
def save_array(file_name_prefix: str, stage: int, metric_name: str, np_array: array):
    filename = f"{file_name_prefix}_{stage}_{metric_name}.array_{np_array.dtype}"
    np_array.tofile(open(filename, "wb"))
    print(f"Saved '{filename}'")

def load_array(file_name_prefix: str, stage: int, metric_name: str, dtype):
        
    filename = f"{file_name_prefix}_{stage}_{metric_name}.array_{dtype}"
    file_size = os.stat(filename).st_size
    
    np_array = np.fromfile(open(filename, "rb"), dtype)
    print(f"Loaded '{filename}'")
    
    return np_array
```

```python
def read_interval(prefix, stage, nbr_samples):
    interval = array('d')
    interval.fromfile(open(f"{prefix}_{stage}_interval.array", "rb"), nbr_samples)
    return interval
```

```python
def get_xy(m, get_samples):
    x = array('d')
    y = array('d')
    
    for _ in range(m):
        samples = get_samples()  #   gamma.rvs(*gamma_params, size=k)
        x.append(sum(samples))
        y.append(max(samples))
        
    return x, y
```

```python
def mask_it(pct, x, y):
    limit = np.percentile(x, pct)
    mask = np.ma.masked_array(x < limit)
    
    x = np.ma.masked_array(x, mask)
    y = np.ma.masked_array(y, mask)

    return x, y
```

```python
def incremental_rt2(initial_fitness, cut_off_growth_chance):
    assert(0 < initial_fitness < 1)
    degree = 1.0
    growth_chance = initial_fitness
    edge_count = count(int(degree / initial_fitness))
    
    while growth_chance > cut_off_growth_chance:
        degree += random() < growth_chance
        growth_chance = degree / next(edge_count) / 2
        # yield (growth_chance, degree, edge_count)
        
    return degree

def generate_incremental_rt(m, initial_fitness, cut_off_growth_chance):
    data = array('d')
    
    for _ in range(m):
        data.append(incremental_rt2(initial_fitness, cut_off_growth_chance, ))
        
    return data
```

```python
def incremental_rt3(initial_fitness, cut_off_growth_chance, ):
    assert(0 < initial_fitness < 1)
    degree = 1.0
    growth_chance = initial_fitness
    edge_count = count(int(degree / initial_fitness))
    
    while growth_chance > cut_off_growth_chance:
        degree += random() * growth_chance  # CHANGED HERE
        growth_chance = degree / next(edge_count) / 2
        # yield (growth_chance, degree, edge_count)
        
    return degree

def generate_incremental_rt3(m, initial_fitness, cut_off_growth_chance, ):
    data = array('d')
    
    for _ in range(m):
        data.append(incremental_rt3(initial_fitness, cut_off_growth_chance, ))
        
    return data
```

```python
def incremental_rt4(initial_fitness, cut_off_growth_chance):
    assert(0 < initial_fitness < 1)
    degree = uniform(0.5, 1.5)
    growth_chance = initial_fitness
    edge_count = count(int(degree / initial_fitness))
    
    while growth_chance > cut_off_growth_chance:
        degree += (random() < growth_chance) * uniform(0.5, 1.5)  # CHANGED HERE
        growth_chance = degree / next(edge_count) / 2
        # yield (growth_chance, degree, edge_count)
        
    return degree

def generate_incremental_rt4(m, initial_fitness, cut_off_growth_chance):
    data = array('d')
    
    for _ in range(m):
        data.append(incremental_rt4(initial_fitness, cut_off_growth_chance))
        
    return data
```

```python
def incremental_rt5(initial_fitness, max_steps):
    assert(0 < initial_fitness < 1)
    degree = initial_fitness
    
    for num_edges in range(1, max_steps):
        if random() < degree / num_edges / 2:
            degree += uniform(0.5, 1.5)
        
    return degree

def generate_incremental_rt5(m, initial_fitness, max_steps):
    data = array('d')
    
    for _ in range(m):
        data.append(incremental_rt5(initial_fitness, max_steps))
        
    return data
```

```python
def incremental_rt6(max_steps):
    degree = 1
    
    for total_degrees in range(2, 2*max_steps, 2):
        limit = degree  / total_degrees / 2
        if limit < 0.00001: break
        if random() < limit:
            degree += 1
    return degree 

def generate_incremental_rt6(m, max_steps):
    data = array('Q')
    
    for _ in range(m):
        data.append(incremental_rt6(max_steps))
        
    rts = array('Q')
    nbr_occurances = array('Q')
    
    for rt, occurances in groupby(np.sort(data)):
        rts.append(rt)
        nbr_occurances.append(len(list(occurances)))
        
    return rts, nbr_occurances
```

```python
m = 300000
with timing(f"Generating {m} data points"):
    data = generate_incremental_rt6(m, max_steps=3000)
    
    
save_array("incremental", 10, "response-time", np.array(data))
```

```python
dNetwork = {0:0, 1:1}       # dictionary of ints
nbr_nodes = 10000       # total number of nodes
total_links = 1

for node_num in range(2, nbr_nodes):
    # initialize node of key i with empty list
    dNetwork[node_num] = 0      
    
    # Add links to the new node from some of the pre-existing nodes
    for other_node_num, other_node_nbr_out_edges in dNetwork.items():
        
        # Propability of getting a link from the current node
        # Proportional to the nbr of pre-existing out-edges of the current node
        fThresh = (other_node_nbr_out_edges ) / (total_links + node_num ) # 
        if(random() <= fThresh):
            dNetwork[other_node_num] += 1
            total_links += 1

# calculate the degree distribution
lDegrees = sorted(dNetwork.values())

# print([ (k, sum(values)) for k, values in groupby(lDegrees)] )
# print()
degrees = tuple(zip(*((k+1, len(list(values))) for k, values in groupby(lDegrees) ) ))
print (degrees[:10])

# and draw a histogram
plt.scatter(*degrees)
plt.xlabel("Node Degree")
plt.ylabel("Number of Nodes")
plt.xscale('log')
plt.yscale('log')
plt.title("Preferential Attachment")
plt.show()

plt.hist(lDegrees,50)
plt.xlabel("Node Degree")
plt.ylabel("Number of Nodes")
plt.xscale('log')
plt.yscale('log')
plt.title("Preferential Attachment")
plt.show()
```
